```python
PATH = "7282_1.csv"
OUT = "7282_1_dedupe.csv"

df = pd.read_csv(PATH)

# Boolean mask: True where the whole row is a duplicate of a previous row
mask = df.duplicated()

df_clean = df.drop_duplicates(keep='first').reset_index(drop=True)
df_clean.to_csv(OUT, index=False)
print("Wrote:", OUT)
```
